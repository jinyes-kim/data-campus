```
model = gensim.models.Doc2Vec.load("./doc2vec_kr.model")
```

```
tmp = model.most_similar(u'닝', topn=20)
tmp2 = model.most_similar(u'소스', topn=20)
tmp3 = model.most_similar(u'드레싱', topn=20)
tmp4 = model.most_similar(positive=[u'과자', u'닝'], topn=20)
```

```
print('\n 시즈닝과 관련된 단어 유사도\n')
res1 = pd.DataFrame(tmp, columns=['단어', '유사도'])
res2 = pd.DataFrame(tmp2, columns=['단어', '유사도'])
res3 = pd.DataFrame(tmp3, columns=['단어', '유사도'])
res4 = pd.DataFrame(tmp4, columns=['단어', '유사도'])
```
